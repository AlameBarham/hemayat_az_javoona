```python
df.ParentalEducation.fillna("Non", inplace=True)
df.ParentalSupport.fillna("Non", inplace=True)

X_test.ParentalEducation.fillna("Non", inplace=True)
X_test.ParentalSupport.fillna("Non", inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
df[["ParentalEducation", "ParentalSupport", "Ethnicity"]] = encoder.fit_transform(df[["ParentalEducation", "ParentalSupport", "Ethnicity"]])
X_test[["ParentalEducation", "ParentalSupport", "Ethnicity"]] = encoder.transform(X_test[["ParentalEducation", "ParentalSupport", "Ethnicity"]])
```

```python
corr_mat = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_mat, annot=True, cmap='coolwarm', linewidths=.5) 
plt.show()
```

## Train_Test_Split

```python
# Train & validation
from sklearn.model_selection import train_test_split

X = df.drop(["StudentID", "GPA"], axis=1)
y = df["GPA"]
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=56)
X_train.isna().sum().sum()
```

```python
# Test

X_test.drop(["StudentID"], axis=1, inplace=True)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge

linear_reg = make_pipeline(
    StandardScaler(),
    Ridge(alpha=50)
)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

linear_reg.fit(X_train, y_train)

print("Train_scores:")
print(f"r2_score: {r2_score(y_train, linear_reg.predict(X_train))}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_train, linear_reg.predict(X_train)))}")

print("Test_scores:")
print(f"r2_score: {r2_score(y_val, linear_reg.predict(X_val))}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_val, linear_reg.predict(X_val)))}")
```

# Ensemble_tree

```python
from sklearn.ensemble import RandomForestRegressor

random_forest = RandomForestRegressor(random_state=56, n_estimators=100, min_samples_split=5)
random_forest.fit(X_train, y_train)

print("Train_scores:")
print(f"r2_score: {r2_score(y_train, random_forest.predict(X_train))}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_train, random_forest.predict(X_train)))}")

print("Test_scores:")
print(f"r2_score: {r2_score(y_val, random_forest.predict(X_val))}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_val, random_forest.predict(X_val)))}")
```

```python
import zipfile
import os

if not os.path.exists(os.path.join(os.getcwd(), 'Student_GPA.ipynb')):
    %notebook -e Student_GPA.ipynb

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

submission.to_csv('submission.csv', index=False)

file_names = ['Student_GPA.ipynb', 'submission.csv']
compress(file_names)
```
